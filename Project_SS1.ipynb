# Maxwell-Boltzmann Distribution 

$f(x)=exp((E_f-E)/k_Bt)$

if $E=0$,$f(x)=exp((E_f)/k_Bt$
if $E=\inf$,$f(x)=0$
if $E=E_f$,$f(x)=1$

```python
#  program for 'Maxwell-Boltzmann Distribution' plot
import numpy as np
import pandas as pd  # import should be used bofore the used of the term
import matplotlib.pyplot as plt
xf=5
kb=1.38E-23
ee  =1.6E-19
#tt=10
X = [xx for xx in range(10)]
M112={"exp(ef-ee), at t=5 K":[(np.exp((xf-xx)*ee*1E-4/  (kb*5)))for  xx in X],\
     "exp(ef-ee),  at t=50 K":[(np.exp((xf-xx)*ee*1E-4/  (kb*50)))for  xx in X],\
     "exp(ef-ee), at t=100 K":[(np.exp((xf-xx)*ee*1E-4/  (kb*100)))for  xx in X]}
#,\"xx=xf":[(xf=xx) for  xx in X]}
DF112=pd.DataFrame(M112)
DF112
DF112.plot ()
plt.xlabel("energy (x0.1 meV)")
#plt.xscale('exp')
#plt.title('linear')
plt.ylabel("Y=exp(ef-ee)")
plt.title('Maxwell-Boltamann Distribution  curve')
```

# Fermi-Dirac Distribution

$f(x)=1/(exp((E-E_f)/k_Bt)+1)$

if $E=0$,$f(x)=1/(exp((E_f)/k_Bt+1)$
if $E=E_f$,$f(x)=1/2$
if $E=\inf$,$f(x)=0$\

```python
#  program for 'Fermi-Dirac Distribution' plot
import numpy as np
import pandas as pd  # import should be used bofore the used of the term
import matplotlib.pyplot as plt
xf=5
kb=1.38E-23
ee  =1.6E-19
X = [xx for xx in range(20)]
M113={"1/1+exp(ee-ef), at t=50 K":[1/(1+(np.exp((xx-xf)*ee*1E-2/  (kb*50))))for  xx in X],\
     "1/1+exp(ee-ef),  at t=100 K":[1/(1+(np.exp((xx-xf)*ee*1E-2/  (kb*100))))for  xx in X],\
     "1/1+exp(ee-ef), at t=300 K":[1/(1+(np.exp((xx-xf)*ee*1E-2/  (kb*300))))for  xx in X]}
DF113=pd.DataFrame(M113)
DF113
DF113.plot ()
plt.xlabel("energy (x10 meV)")
plt.ylabel("Y=1/1+exp(ee-ef)")
plt.title('Fermi-Dirac Distribution  curve')
```

# Bose-Einstein Distribution

if $E=0$,$f(x)=1/(exp^(-E_f)/k_Bt)-1)$
if $E=E_f$,$f(x)=inf$
if $E<=E_f$,$f(x)=-1$
if $E>=E_f$, $f(x)=0$

```python
#  program for 'Bose-Einstein Distribution' plot
import numpy as np
import pandas as pd  # import should be used bofore the used of the term
import matplotlib.pyplot as plt
xf=10
kb=1.38E-23
ee  =1.6E-19
#tt=10
X = [xx for xx in range(20)]
M114={"1/exp(ee-ef)-1, at t=50 K":[1/(-1+(np.exp((xx-xf)*ee*1E-3/  (kb*50))))for  xx in X],\
     "1/exp(ee-ef)-1,  at t=100 K":[1/(-1+(np.exp((xx-xf)*ee*1E-3/  (kb*100))))for  xx in X],\
     "1/exp(ee-ef)-1, at t=300 K":[1/(-1+(np.exp((xx-xf)*ee*1E-3/  (kb*300))))for  xx in X]}
DF114=pd.DataFrame(M114)
DF114
DF114.plot ()
plt.xlabel("energy ( meV)")
plt.ylabel("Y=1/exp(ee-ef)-1")
plt.title('Bose-Eisntein Distribution  curve')
```
